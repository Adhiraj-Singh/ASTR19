```python
import matplotlib.pyplot as plt
import math
import numpy as np
```

```python
x = np.linspace(0,1,100)
```

```python
print(x)
```

```python
def sinVal(x):
    arr=[]
    for z in x:
        arr.append(math.sin(z))
    return arr

def cosVal(x):
    arr=[]
    for z in x:
        arr.append(math.cos(z))
    return arr
```

```python
sinValArr = sinVal(x)
cosValArr = cosVal(x)
```

```python
fig, (axis1, axis2) = plt.subplots(1,2)
fig.suptitle("Sin(x)                   Cos(x)")
axis1.plot(sinValArr,x)
axis2.plot(cosValArr,x)

plt.show()
```
